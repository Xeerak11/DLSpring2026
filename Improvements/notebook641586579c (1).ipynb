## What this notebook does

We inject a **VLM surprise gate** into the DreamerV3 training loop.  
At every environment step our `VLMGymWrapper` computes a pixel-level frame-difference  
score as a **real-time proxy for DreamerV3's dynamics loss**. When this score spikes  
(rolling z-score > threshold), it calls **Gemini 1.5 Flash** with a screenshot + context,  
parses the returned action plan, and queues those actions to override the agent for the next N steps.

### ❓ Does DreamerV3 already handle the CNN?
**YES.** DreamerV3 contains a full CNN encoder inside its RSSM world model. It processes  
raw pixel frames and learns feature representations end-to-end. **You do not need a separate CNN.**  
The dynamics loss (`train/loss/dyn`) is already computed and logged internally by DreamerV3.  
We use pixel MSE as a real-time proxy because DreamerV3's JAX-compiled training loop  
cannot be paused mid-step for Python callbacks.

```python
%%writefile /kaggle/working/vlm_gate.py
import os, io, json, warnings
import numpy as np
from collections import deque

# Suppress google.generativeai deprecation warnings to prevent silent I/O crashes
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", message=".*google.generativeai.*")

try:
    import google.generativeai as genai
    _GENAI_AVAILABLE = True
except ImportError:
    _GENAI_AVAILABLE = False

ACTION_NAMES = {
    0:"NOOP", 1:"FIRE", 2:"UP", 3:"RIGHT", 4:"LEFT", 5:"DOWN",
    6:"UPRIGHT", 7:"UPLEFT", 8:"DOWNRIGHT", 9:"DOWNLEFT", 10:"UPFIRE",
    11:"RIGHTFIRE", 12:"LEFTFIRE", 13:"DOWNFIRE", 14:"UPRIGHTFIRE",
    15:"UPLEFTFIRE", 16:"DOWNRIGHTFIRE", 17:"DOWNLEFTFIRE",
}
NAME_TO_ACTION = {v: k for k, v in ACTION_NAMES.items()}
VALID_VLM_NAMES = ["NOOP","FIRE","UP","RIGHT","LEFT","DOWN",
                   "UPRIGHT","UPLEFT","DOWNRIGHT","DOWNLEFT"]

SITUATION_REWARD = {
    "key": 1.5, "new room": 1.2, "door": 1.0, "ladder": 0.6,
    "rope": 0.5, "treasure": 0.8, "gem": 0.8, "amulet": 0.8,
    "sword": 0.8, "torch": 0.5, "progress": 0.4, "climbing": 0.3,
    "moving": 0.2, "collecting": 0.5,
    "skull": -0.5, "enemy": -0.4, "snake": -0.5, "spider": -0.5,
    "dead": -1.0, "died": -1.0, "stuck": -0.3, "falling": -0.2,
    "pit": -0.4, "danger": -0.3,
}

def _parse_reward_from_description(description):
    desc_lower = description.lower()
    bonus = 0.0
    for keyword, value in SITUATION_REWARD.items():
        if keyword in desc_lower:
            bonus += value
    return float(np.clip(bonus, -2.0, 2.0))

def _diverse(actions):
    if not actions:
        return False
    noop_count = sum(1 for a in actions if a == 0)
    return noop_count < len(actions) * 0.7

class VLMGate:
    def __init__(self, api_key="", threshold_z=0.5, threshold_z_start=0.5,
                 anneal_steps=100_000, queue_len=25,
                 cooldown_steps=60, reward_scale=0.3, enabled=True,
                 log_path="/kaggle/working/vlm_call_log.json"):
        self.threshold_z        = threshold_z
        self.threshold_z_start  = threshold_z_start
        self.anneal_steps       = anneal_steps
        self.queue_len          = queue_len
        self.cooldown_steps     = cooldown_steps
        self.reward_scale       = reward_scale
        self.enabled            = enabled
        self.log_path           = log_path

        self._queue               = deque()
        self._prev_frame          = None
        self._surprise_buf        = deque(maxlen=200)
        self._cooldown            = 0
        self._active_reward_bonus = 0.0
        self._visit_counts        = {}   

        self.total_vlm_calls = 0
        self.total_steps     = 0
        self._call_log       = []

        self._model = None
        if enabled and api_key and _GENAI_AVAILABLE:
            genai.configure(api_key=api_key)
            self._model = genai.GenerativeModel("gemini-1.5-flash")
            print(f"[VLMGate] Gemini 1.5 Flash ready "
                  f"threshold_z={threshold_z} (start={threshold_z_start}) "
                  f"queue_len={queue_len} cooldown={cooldown_steps} "
                  f"reward_scale={reward_scale}")

    def _current_threshold(self):
        t = min(1.0, self.total_steps / max(1, self.anneal_steps))
        return self.threshold_z_start + t * (self.threshold_z - self.threshold_z_start)

    def _surprise(self, frame):
        frame_f = frame.astype(np.float32)
        if self._prev_frame is None:
            self._prev_frame = frame_f
            return 0.0
        mse = float(np.mean((frame_f - self._prev_frame) ** 2))
        self._prev_frame = frame_f
        return mse

    def _should_trigger(self, surprise):
        self._surprise_buf.append(surprise)
        if len(self._surprise_buf) < 20:
            return False
        arr = np.array(self._surprise_buf)
        z   = (surprise - arr.mean()) / (arr.std() + 1e-6)
        return bool(z > self._current_threshold())

    def _curiosity_bonus(self, frame):
        try:
            small = frame.astype(np.float32)
            if small.ndim == 3:
                small = small.mean(axis=-1)
            h, w = small.shape
            bh, bw = max(1, h // 8), max(1, w // 8)
            bucket = tuple(
                int(small[r*bh:(r+1)*bh, c*bw:(c+1)*bw].mean() // 16)
                for r in range(8) for c in range(8)
            )
            count = self._visit_counts.get(bucket, 0) + 1
            self._visit_counts[bucket] = count
            return float(1.0 / np.sqrt(count)) * 0.05
        except Exception:
            return 0.0

    def _call_vlm(self, frame, step, score, lives, hist, is_episode_start=False):
        if self._model is None:
            return [], 0.0, ""
        try:
            from PIL import Image
            frame_u8 = np.clip(frame, 0, 255).astype(np.uint8)
            if frame_u8.ndim == 2:
                pil = Image.fromarray(frame_u8, "L").convert("RGB")
            elif frame_u8.shape[-1] == 1:
                pil = Image.fromarray(frame_u8[:, :, 0], "L").convert("RGB")
            else:
                pil = Image.fromarray(frame_u8)
            pil = pil.resize((320, 240), Image.NEAREST)
            buf = io.BytesIO()
            pil.save(buf, format="PNG")
            img_bytes = buf.getvalue()

            hist_names = [ACTION_NAMES.get(int(a), "NOOP") for a in hist[-6:]]
            valid_str  = str(VALID_VLM_NAMES)
            n_actions  = self.queue_len
            
            context_prefix = "The agent is starting a new episode. Chart an optimal path.\n" if is_episode_start else "The agent has encountered a surprising situation and needs guidance.\n"

            prompt = (
                "You are an expert player of Montezuma's Revenge (Atari 2600).\n"
                f"{context_prefix}\n"
                f"Current state: step={step}  score={score:.0f}  lives={lives}\n"
                f"Last actions:  {hist_names}\n\n"
                "Examine the screenshot carefully. Respond with ONLY a JSON object -- "
                "no markdown, no backticks, no explanation.\n\n"
                "Required fields:\n"
                '  \"description\": 1-2 sentences mentioning keys, enemies, ladders, '
                "rooms, ropes, platforms, and whether the agent is making progress.\n"
                f'  \"actions\": array of exactly {n_actions} strings chosen from {valid_str}.\n'
                '  \"reward_bonus\": float in [-2.0, 2.0]. '
                "Positive = good position or progress, negative = near enemies / stuck / dying.\n\n"
                "Strategy tips:\n"
                "- Collect keys before unlocking doors.\n"
                "- Avoid skulls and snakes; use ropes and ladders to navigate.\n"
                "- Prefer RIGHT/LEFT/UP/DOWN over NOOP when progress is possible.\n\n"
                'Example: {\"description\": \"Agent is starting near the ladders.\", '
                '\"actions\": [\"RIGHT\",\"UP\",\"RIGHT\",\"RIGHT\",\"UP\",'
                '\"LEFT\",\"LEFT\",\"RIGHT\",\"FIRE\",\"NOOP\"], '
                '\"reward_bonus\": 1.2}'
            )

            response = self._model.generate_content([
                {"mime_type": "image/png", "data": img_bytes},
                prompt,
            ])
            
            # FIX: Dynamically generate triple backticks to avoid Kaggle notebook parser crashing
            raw = response.text.strip()
            raw = raw.replace('`' * 3 + 'json', '').replace('`' * 3, '').strip()
            parsed = json.loads(raw)

            description = str(parsed.get("description", ""))
            raw_actions = parsed.get("actions", [])
            raw_bonus   = parsed.get("reward_bonus", None)

            ints = [NAME_TO_ACTION.get(str(n).upper().strip(), 0) for n in raw_actions]
            ints = ints[:self.queue_len]

            if not _diverse(ints):
                print(f"[VLMGate] step={step} -- low-diversity actions from VLM, skipping.")
                return [], 0.0, description

            if raw_bonus is not None:
                try:
                    reward_bonus = float(np.clip(float(raw_bonus), -2.0, 2.0))
                except (TypeError, ValueError):
                    reward_bonus = _parse_reward_from_description(description)
            else:
                reward_bonus = _parse_reward_from_description(description)

            return ints, reward_bonus, description

        except Exception as e:
            print(f"[VLMGate] call failed: {e}")
            return [], 0.0, ""

    def get_action(self, frame, agent_action, step=0, score=0.0,
                   lives=3, action_history=None, is_episode_start=False):
        
        if not self.enabled:
            return agent_action, False, 0.0

        self.total_steps += 1
        action_history = action_history or []

        curiosity = self._curiosity_bonus(frame)

        if self._queue:
            vlm_scaled = self._active_reward_bonus * self.reward_scale
            return self._queue.popleft(), True, vlm_scaled + curiosity

        self._active_reward_bonus = 0.0

        surprise = self._surprise(frame)
        if self._cooldown > 0 and not is_episode_start:
            self._cooldown -= 1
            return agent_action, False, curiosity

        if is_episode_start or self._should_trigger(surprise):
            vlm_acts, reward_bonus, description = self._call_vlm(
                frame, step, score, lives, action_history, is_episode_start)
            if vlm_acts:
                self._queue.extend(vlm_acts)
                self._active_reward_bonus = reward_bonus
                self._cooldown = self.cooldown_steps
                self.total_vlm_calls += 1
                self._call_log.append({
                    "step":         step,
                    "surprise":     round(surprise, 2),
                    "threshold":    round(self._current_threshold(), 3),
                    "reward_bonus": round(reward_bonus, 3),
                    "description":  description,
                    "actions":      [ACTION_NAMES.get(a, "NOOP") for a in vlm_acts],
                    "is_start":     is_episode_start
                })
                act_names = [ACTION_NAMES.get(a, "?") for a in vlm_acts[:4]]
                print(f"[VLMGate] step={step:>8}  surprise={surprise:7.2f}  "
                      f"thr={self._current_threshold():.2f}  "
                      f"reward_bonus={reward_bonus:+.2f}  -> {act_names}...")
                vlm_scaled = self._active_reward_bonus * self.reward_scale
                return self._queue.popleft(), True, vlm_scaled + curiosity

        return agent_action, False, curiosity

    def reset_episode(self):
        self._queue.clear()
        self._prev_frame          = None
        self._cooldown            = 0
        self._active_reward_bonus = 0.0

    def save_log(self):
        data = {
            "total_steps":     self.total_steps,
            "total_vlm_calls": self.total_vlm_calls,
            "call_pct":        round(100 * self.total_vlm_calls / max(1, self.total_steps), 3),
            "unique_states":   len(self._visit_counts),
            "calls":           self._call_log,
        }
        with open(self.log_path, "w") as fh:
            json.dump(data, fh, indent=2)

    def stats(self):
        return {
            "total_steps":       self.total_steps,
            "total_vlm_calls":   self.total_vlm_calls,
            "call_pct":          f"{100*self.total_vlm_calls/max(1,self.total_steps):.2f}%",
            "threshold_z":       self.threshold_z,
            "threshold_z_start": self.threshold_z_start,
            "queue_len":         self.queue_len,
            "cooldown_steps":    self.cooldown_steps,
            "reward_scale":      self.reward_scale,
            "unique_states":     len(self._visit_counts),
        }
print("Written: /kaggle/working/vlm_gate.py")
```

```python
%%writefile /kaggle/working/vlm_gym_wrapper.py
import sys
import numpy as np

sys.path.insert(0, "/kaggle/working")
from vlm_gate import VLMGate

class VLMGymWrapper:
    def __init__(self, env, api_key="", threshold_z=0.5, threshold_z_start=0.5,
                 anneal_steps=100_000, queue_len=25,
                 cooldown_steps=60, reward_scale=0.3):
        self.env  = env
        self.gate = VLMGate(
            api_key           = api_key,
            threshold_z       = threshold_z,
            threshold_z_start = threshold_z_start,
            anneal_steps      = anneal_steps,
            queue_len         = queue_len,
            cooldown_steps    = cooldown_steps,
            reward_scale      = reward_scale,
            enabled           = True,
            log_path          = "/kaggle/working/vlm_call_log.json",
        )
        self._step_count            = 0
        self._score                 = 0.0
        self._lives                 = 3
        self._action_hist           = []
        self._current_obs           = None
        self._is_first_step         = True
        self._prev_ram_inventory    = 0
        self.total_vlm_reward_added = 0.0
        self.total_ram_bonus_added  = 0.0
        self.total_reward_steps     = 0

        for attr in ("observation_space", "action_space", "reward_range", "metadata", "spec"):
            try:
                setattr(self, attr, getattr(env, attr))
            except AttributeError:
                pass

    def _curriculum_scale(self):
        n_halvings = self._step_count // 200_000
        return self.gate.reward_scale * (0.5 ** n_halvings)

    def reset(self, **kwargs):
        result = self.env.reset(**kwargs)
        if isinstance(result, tuple) and len(result) == 2:
            obs, info = result
        else:
            obs, info = result, {}
        self._current_obs = np.array(obs)
        self.gate.reset_episode()
        self._score       = 0.0
        self._lives       = 3
        self._action_hist = []
        self._is_first_step = True
        self._prev_ram_inventory = 0
        
        if isinstance(result, tuple) and len(result) == 2:
            return obs, info
        return obs

    def step(self, action):
        frame = (self._current_obs if self._current_obs is not None
                 else np.zeros((210, 160, 3), np.uint8))

        is_start = self._is_first_step
        self._is_first_step = False

        orig_scale = self.gate.reward_scale
        self.gate.reward_scale = self._curriculum_scale()

        final_action, was_override, reward_bonus = self.gate.get_action(
            frame          = frame,
            agent_action   = int(action),
            step           = self._step_count,
            score          = self._score,
            lives          = self._lives,
            action_history = self._action_hist[-6:],
            is_episode_start = is_start
        )

        self.gate.reward_scale = orig_scale  

        result = self.env.step(final_action)
        if len(result) == 5:
            obs, env_reward, terminated, truncated, info = result
            done = terminated or truncated
        else:
            obs, env_reward, done, info = result

        # RAM-Based Key Detection (+50 Hard Bonus)
        ram_bonus = 0.0
        try:
            if hasattr(self.env, 'unwrapped') and hasattr(self.env.unwrapped, 'ale'):
                ram = self.env.unwrapped.ale.getRAM()
                current_inventory = int(ram[67])
                if current_inventory > self._prev_ram_inventory:
                    ram_bonus = 50.0  
                    self.total_ram_bonus_added += ram_bonus
                    print(f"[VLMGymWrapper] RAM KEY DETECTED! +50 Bonus Added (step={self._step_count})")
                self._prev_ram_inventory = current_inventory
        except Exception:
            pass

        shaped_reward = float(env_reward) + reward_bonus + ram_bonus
        if reward_bonus != 0.0 or ram_bonus != 0.0:
            self.total_vlm_reward_added += reward_bonus
            self.total_reward_steps     += 1

        self._current_obs  = np.array(obs)
        self._score       += float(env_reward)
        self._step_count  += 1
        self._action_hist.append(int(final_action))
        if isinstance(info, dict) and "lives" in info:
            self._lives = int(info["lives"])

        return obs, shaped_reward, done, info

    def __getattr__(self, name):
        return getattr(self.env, name)

    def close(self):
        self.gate.save_log()
        self.env.close()

    def render(self, *a, **kw):
        return self.env.render(*a, **kw)

    def seed(self, seed=None):
        if hasattr(self.env, "seed"):
            return self.env.seed(seed)
print("Written: /kaggle/working/vlm_gym_wrapper.py")
```

```python
# Cell 12 — load metrics from both runs
import json, glob, os
import numpy as np

LOGDIR_BASE = "/kaggle/working/dreamer_montezuma_400k"   # Deliverable 3
LOGDIR_VLM  = "/kaggle/working/dreamer_montezuma_vlm"    # This notebook

def load_metrics(logdir):
    """Load metrics.jsonl from a dreamerv3 logdir."""
    files = glob.glob(os.path.join(logdir, "**", "metrics.jsonl"), recursive=True)
    if not files:
        files = glob.glob(os.path.join(logdir, "*.jsonl"), recursive=True)
    if not files:
        return []
    records = []
    with open(files[0]) as f:
        for line in f:
            line = line.strip()
            if line:
                try:
                    records.append(json.loads(line))
                except Exception:
                    pass
    return records

def extract(records, candidates):
    steps, vals = [], []
    for r in records:
        step = r.get("step", r.get("steps"))
        if step is None:
            continue
        for k in candidates:
            if k in r and isinstance(r[k], (int, float)):
                steps.append(step)
                vals.append(r[k])
                break
    return np.array(steps), np.array(vals)

SCORE_KEYS = ["episode/score", "train/episode_score", "episode_score", "return", "reward"]
DYN_KEYS   = ["train/loss/dyn", "dyn_loss", "loss/dyn"]

base_records = load_metrics(LOGDIR_BASE)
vlm_records  = load_metrics(LOGDIR_VLM)

base_s_steps, base_scores = extract(base_records, SCORE_KEYS)
vlm_s_steps,  vlm_scores  = extract(vlm_records,  SCORE_KEYS)
base_d_steps, base_dyn    = extract(base_records, DYN_KEYS)
vlm_d_steps,  vlm_dyn     = extract(vlm_records,  DYN_KEYS)

print(f"Baseline  : {len(base_scores)} score entries, {len(base_dyn)} dyn-loss entries")
print(f"VLM-gated : {len(vlm_scores)} score entries, {len(vlm_dyn)} dyn-loss entries")

if len(base_scores) > 0:
    print(f"  baseline  max score : {base_scores.max():.1f}  mean : {base_scores.mean():.1f}")
if len(vlm_scores) > 0:
    print(f"  vlm-gated max score : {vlm_scores.max():.1f}  mean : {vlm_scores.mean():.1f}")
```

```python
# Cell 18 — Plot 4: Reward shaping, curiosity, and adaptive threshold analysis

import matplotlib.pyplot as plt
import numpy as np

calls = vlm_log.get("calls", [])

fig, axes = plt.subplots(1, 3, figsize=(17, 5))
fig.suptitle("Reward Shaping & Curiosity Analysis", fontsize=13)

# subplot 1: reward bonus distribution
ax = axes[0]
bonuses = [c.get("reward_bonus", 0) for c in calls]
if bonuses:
    ax.hist(bonuses, bins=20, color="#7F77DD", edgecolor="white", alpha=0.85)
    ax.axvline(0, color="red", linestyle="--", linewidth=1)
    ax.axvline(np.mean(bonuses), color="orange", linestyle="-",
               linewidth=1.5, label=f"mean={np.mean(bonuses):+.2f}")
    ax.set_xlabel("Reward Bonus per VLM Call", fontsize=11)
    ax.set_ylabel("Count", fontsize=11)
    ax.set_title("VLM Reward Bonus Distribution", fontsize=11)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)
else:
    ax.text(0.5, 0.5, "No data\n(run training first)",
            transform=ax.transAxes, ha="center", va="center", fontsize=11)

# subplot 2: adaptive threshold over time
ax2 = axes[1]
thr_steps = [c["step"]      for c in calls if "threshold" in c]
thr_vals  = [c["threshold"] for c in calls if "threshold" in c]
if thr_steps:
    ax2.scatter(thr_steps, thr_vals, s=18, color="#F5A623", alpha=0.7, zorder=3)
    ax2.set_xlabel("Environment Step", fontsize=11)
    ax2.set_ylabel("Active z-score Threshold", fontsize=11)
    ax2.set_title("Adaptive Threshold Annealing", fontsize=11)
    ax2.grid(True, alpha=0.3)
else:
    ax2.text(0.5, 0.5, "No threshold data\n(re-run after training)",
             transform=ax2.transAxes, ha="center", va="center", fontsize=11)

# subplot 3: cumulative VLM calls
ax3 = axes[2]
total_steps = vlm_log.get("total_steps", 0)
if calls and total_steps > 0:
    call_steps = sorted(c["step"] for c in calls)
    cum_calls  = list(range(1, len(call_steps) + 1))
    ax3.plot(call_steps, cum_calls, color="#378ADD", linewidth=2)
    ax3.fill_between(call_steps, cum_calls, alpha=0.2, color="#378ADD")
    ax3.set_xlabel("Environment Step", fontsize=11)
    ax3.set_ylabel("Cumulative VLM Calls", fontsize=11)
    ax3.set_title("Cumulative VLM Call Rate", fontsize=11)
    ax3.grid(True, alpha=0.3)
    ax3.text(0.02, 0.96,
             f"total: {len(calls)}  rate: {vlm_log['call_pct']:.3f}%",
             transform=ax3.transAxes, va="top", fontsize=9,
             bbox=dict(boxstyle="round", facecolor="lightyellow", alpha=0.8))
else:
    ax3.text(0.5, 0.5, "No data", transform=ax3.transAxes,
             ha="center", va="center", fontsize=11)

plt.tight_layout()
plt.savefig("/kaggle/working/cmp_plot4_reward_shaping.png", dpi=150, bbox_inches="tight")
plt.show()
print("saved: cmp_plot4_reward_shaping.png")
```
